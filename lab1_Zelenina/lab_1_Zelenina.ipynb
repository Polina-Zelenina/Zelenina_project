### Робота з віддаленим репозиторієм

Будь-які зміни в проєкті можуть бути додані в індекс Git-репозиторію і закомічені у репозиторій на локальній машині. Після того копія локального репо може бути записана у віддалений репо на GitHub.
Перелік необхідних команд Git, якщо викоростовувати для цього консоль, може бути таким:

* `git status` - перевірка статусу файлів;

* `git add .` - додавання всіх новостворених файлів у індекс;

* `git commit -m "коментар до коміту"` - додавання змін у локальний Git-репозиторій;

* `git push -u origin master` - відсилання поточної локальної копії Git-репо на GitHub-репозиторій.

Початківці це можуть зробити простіше через меню PyCharm.

## Контрольні запитання

1. У чому відмінність інтерпретатора від компілятора? Python -- компілятор чи інтерпретатор?
   * Компіляція - вихідна програма транслюється один раз (проте цю дію необхідно повторювати щоразу, коли ви змінюєте вихідний код) шляхом отримання файлу (наприклад, файла .exe, якщо код призначений для запуску в MS Windows), що містить машинний код; програма, яка виконує цей переклад, називається компілятором
   * Інтерпретація - користувач коду може перекладати вихідну програму щоразу, коли вона має бути запущена; програма, що виконує таке перетворення, називається інтерпретатором;кінцевому користувачеві також потрібен інтерпретатор для його виконання.
   * Python -- інтерпретатор.

1. Що таке IDE PyCharm?
   * PyCharm - це інтегроване середовище розробки (IDE) для мови програмування Python, розроблене компанією JetBrains. IDE - це спеціалізоване програмне забезпечення, яке допомагає розробникам створювати, редагувати, тестувати і відлагоджувати програми більш ефективно.
    
1. Що таке система контроля версій (СКВ) і для чого вона призначена? У чому принципова відмінність СКВ SVN та GIT?
   * Система контролю версій (СКВ) - це програмне забезпечення, призначене для відстеження змін у вихідному коді або в будь-яких інших текстових документах. Git став більш популярним і універсальним вибором серед розробників завдяки своїй розподіленій природі, швидкості та потужним інструментам для роботи з гілками.

1. Що відбувається при виконанні команди `git add .`?
   * Додавання всіх новостворених файлів у індекс.

1. Що відбувається при виконанні команди `git commit`?
   * Додавання змін у локальний Git-репозиторій.

1. Якою Git-командою виконується розміщення локальної копії Git-репозиторію проєкта на віддаленому, що знаходиться на GitHub?
   * git push -u origin main
